```python
import pandas as pd
import plotly_express as px
import numpy as np
```

```python
car_data = pd.read_csv("../datasets/vehicles_us.csv")
```

```python
car_data.info()
```

```python
car_data.head()
```

```python
car_data.isna().sum()
```

```python
car_data['manufacturer'] = car_data['model'].apply(lambda x: x.split(' ')[0])
```

```python
not_na_car_data = car_data[~car_data['model_year'].isna()]
```

```python
pivot_manufacturer = car_data.pivot_table(index='manufacturer', columns='type', values='fuel', aggfunc='count')
# pivot_manufacturer = pivot_manufacturer.stack().reset_index()
# pivot_manufacturer.columns = ['manufacturer', 'type', 'count']
pivot_manufacturer
```

```python
fig = px.bar(pivot_manufacturer,
             labels={'manufacturer': 'Fabricante', 'value' :'Cantidad de vehículos'},
             barmode='stack')
fig.update_layout(
    xaxis=dict(
        tickangle=70 
    )
)
fig.show()
```

```python
fig = px.histogram(not_na_car_data,
                   x='model_year',
                   color='condition',
                   labels={'model_year': 'Año de fabricación', 'count' :'Cantidad'},
             )
fig.update_layout(
    height=600, 
)
fig.show()
```
